```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
import re
import pandas as pd

driver = webdriver.Chrome()

# Ouvrir la page Web
driver.get("https://aqicn.org/city/paris/")
```

```python
df = pd.DataFrame(columns = ['Station', 'Date', 'PM 2.5'])
```

```python
elements_to_hover_over = driver.find_elements(By.CSS_SELECTOR, "rect.rect-day")

for element in elements_to_hover_over:
    hover = ActionChains(driver).move_to_element(element)
    hover.perform()

    # Attendre un certain temps pour que les informations apparaissent
    driver.implicitly_wait(5)  # Attendre 5 secondes (ajustez selon vos besoins)

    # Récupérer le contenu de la page après le survol
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    tooltip_element = soup.find("div", class_="yearly-aqi-tooltip")
    # Extraire le contenu de l'élément
    if tooltip_element:
        tooltip_content = tooltip_element.get_text(strip=True)
        aqi_value_match = re.search(r'AQI: (\d+)', tooltip_content)
        aqi_value = aqi_value_match.group(1)
        date = tooltip_content.split("AQI:")[0].strip()
        df2 = {'Station': 'Dijon', 'Date': date, 'PM 2.5': aqi_value} 
        print(df2)
        df = df._append(df2, ignore_index = True) 
    else:
        print("no tooltip")
        if(len(df)>340):
            break
        else:
            continue
```

```python
df.to_csv("station_dijon_2021.csv")
```
